```python
import numpy as np
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/dineshpiyasamara/LaptopPricePredictor/master/model%20building/laptop_price.csv"
data = pd.read_csv(url, encoding= 'latin-1')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.head()
```

```python
data['Ram'] = data['Ram'].str.replace('GB','').astype('int32')
data['Weight'] = data['Weight'].str.replace('kg','').astype('float32')
data.head()
```

```python
data.info()
```

```python
numerical_columns = ['Inches','Ram','Weight' , 'Price_euros']
data[numerical_columns].corr()['Price_euros']
```

```python
# checking Company column
data['Company'].value_counts()
```

```python
def add_company(input):
    if input == 'Samsung' or input == 'Razer' or input == 'Mediacom' or input == 'Microsoft' or input == 'Xiaomi' or input == 'Vero' or input == 'Chuwi' or input == 'Google' or input == 'Fujitsu' or input == 'LG' or input == 'Huawei':
        return 'Other'
    else:
        return input

data['Company'] = data['Company'].apply(add_company)
```

```python
data['Company'].value_counts()
```

```python
# checking Product column
data['Product'].value_counts()
```

```python
len(data['Product'].value_counts())
```

```python
# checking TypeName column
data['TypeName'].value_counts()
```

```python
len(data['TypeName'].value_counts())
```

```python
# checking ScreenResolution column
data['ScreenResolution'].value_counts()
```

```python
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['Ips'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
data.head(2)
```

```python
# checking Cpu column
data['Cpu'].value_counts()
```

```python
len(data['Cpu'].value_counts())
```

```python
data['cpu_name'] = data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
data['cpu_name'].value_counts()
```

```python
def set_processor(name):
    if name == 'Intel Core i7' or name == 'Intel Core i5' or name == 'Intel Core i3':
        return name
    else:
        if name.split()[0] == 'AMD':
            return 'AMD'
        else:
            return 'Other'

data['cpu_name'] = data['cpu_name'].apply(set_processor)
```

```python
data['cpu_name'].value_counts()
```

```python
# checking Gpu column
data['Gpu'].value_counts()
```

```python
data['gpu_name'] = data['Gpu'].apply(lambda x:" ".join(x.split()[0:1]))
```

```python
data['gpu_name'].value_counts()
```

```python
data.shape
```

```python
data = data[data['gpu_name'] != 'ARM']
```

```python
data.shape
```

```python
data.head(2)
```

```python
# checking OpSys column
data['OpSys'].value_counts()
```

```python
def set_os(input):
    if input == 'Windows 10' or input == 'Windows 7' or input == 'Windows 10 S':
        return 'Windows'
    elif input == 'macOS' or input == 'Mac OS X':
        return 'Mac'
    elif input == 'Linux':
        return input
    else:
        return 'Other'

data['OpSys'] = data['OpSys'].apply(set_os)
```

```python
data.head()
```

```python
# dropping irrelavant columns
data = data.drop(columns = ['laptop_ID', 'Inches','ScreenResolution', 'Product', 'Cpu', 'Gpu'])
```

```python
data.head()
```

```python
data.info()
```

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
data.info()
```

```python
# Convert boolean values to integer
boolean_columns = data.select_dtypes(include=bool).columns
data[boolean_columns] = data[boolean_columns].astype(int)
```

```python
data.head(2)
```

```python
data.shape
```

```python
X = data.drop('Price_euros', axis=1)
y = data['Price_euros']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape, X_test.shape
```

```python
def model_accuracy(model):
    model.fit(X_train, y_train)
    accuracy = model.score(X_test, y_test)
    print(str(model) + '--->' + str(accuracy))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_accuracy(lr)

from sklearn.linear_model import Lasso
lasso = Lasso()
model_accuracy(lasso)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_accuracy(dt)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_accuracy(rf)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10, 50, 100],
             'criterion':['squared_error', 'absolute_error', 'poisson']}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)

grid_fit = grid_obj.fit(X_train, y_train)

best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(X_test, y_test)
```

```python
import pickle
with open('predictor.pickle', 'wb') as file:
    pickle.dump(best_model, file)
```
